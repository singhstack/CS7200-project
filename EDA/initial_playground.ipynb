Lets see if the time varies with presence/absence of an algorithm

```python
# Specify the number of rows and columns for subplots
rows = 3
cols = 3
# Set up the subplots
fig, axes = plt.subplots(rows, cols, figsize=(15, 10))

# Flatten the axes for easier indexing
axes = axes.flatten()

# Iterate through each binary column and create a subplot
for i, binary_col in enumerate(df.columns[-9:]):  # Exclude the last column (numerical)
    sns.histplot(data=df, x='Opttime', hue=binary_col, kde=True, multiple='stack', ax=axes[i])
    axes[i].set_title(f'Distribution of numerical for {binary_col}')
    axes[i].legend()

# Adjust layout
plt.tight_layout()
# Show the plot
plt.show()
```

```python
# Specify the number of rows and columns for subplots
rows = 3
cols = 3
# Set up the subplots
fig, axes = plt.subplots(rows, cols, figsize=(15, 10))

# Flatten the axes for easier indexing
axes = axes.flatten()

# Iterate through each binary column and create a subplot
for i, binary_col in enumerate(df.columns[-9:]):  # Exclude the last column (numerical)
    sns.scatterplot(data=df, x=binary_col, hue=binary_col, y='Opttime', ax=axes[i])
    axes[i].set_title(f'Distribution of numerical for {binary_col}')
axes[i].legend()
# Adjust layout
plt.tight_layout()
# Show the plot
plt.show()
```

its more or less the same with or without the algorithm. So Are there other factors that cause the time to be small or large?

```python
#Let's see if variables have anything to do with this
plt.figure(figsize = (6,5))
sns.scatterplot(data=df, x='Total vars', y='Opttime')
plt.show()
```
